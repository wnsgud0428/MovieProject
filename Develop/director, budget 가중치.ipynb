### dataframe ê°€ê³µ

```python
# Parse the stringified features into their corresponding python objects
from ast import literal_eval

features = ['cast', 'crew', 'keywords', 'genres']
for feature in features:
    df2[feature] = df2[feature].apply(literal_eval)
```

```python
def get_director(x):
    for i in x:
        if i['job'] == 'Director':
            return i['name']
    return np.nan
```

```python
def get_list(x):
    if isinstance(x, list):
        names = [i['name'] for i in x]
        #Check if more than 3 elements exist. If yes, return only first three. If no, return entire list.
        if len(names) > 3:
            names = names[:3]
        return names

    #Return empty list in case of missing/malformed data
    return []
```

```python
df2['director'] = df2['crew'].apply(get_director)

features = ['cast', 'keywords', 'genres']
for feature in features:
    df2[feature] = df2[feature].apply(get_list)
```

```python
df2 = df2.dropna(subset=['director'])
```

```python
df2 = df2.dropna(subset=['budget'])
```

### director ê°€ì¤‘ì¹˜ 

```python
director = df2['director'].unique()
```

```python
director_revenue = []
for i in range(len(director)):
    director_revenue.append(df2.loc[df2['director'].str.contains(director[i]), 'revenue'].mean())
```

```python
director_data = pd.DataFrame(director,columns=['director'])
director_data['director_revenue'] = director_revenue
```

```python
director_data['director_revenue'].describe()
```

```python
weight_list = []
for i in range(len(director_data)):
    if director_data['director_revenue'][i] < 9.132000e+06:
        weight_list.append('poor')
    elif 9.132000e+06 <= director_data['director_revenue'][i] <= 3.347285e+07:
        weight_list.append('soso')
    elif 3.347285e+07 < director_data['director_revenue'][i] <= 9.182001e+07:
        weight_list.append('hit')
    else:
        weight_list.append('super-hit')
```

```python
director_data['director_weight'] = weight_list
```

```python
director_data['director_revenue'] = director_data['director_revenue'] *(1/1000000)
```
